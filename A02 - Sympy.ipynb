**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x):
    return (1260+117*x-9*x**2)
df=sp.diff(f(x), x)
pc=sp.solve(df, x)

ddf=sp.diff(df, x)

print("El numero crítico se encuentra en " + str(pc))
print("La segunda derivada es" + str(ddf))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(-200,200,500)

plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    plt.plot(c, f(c), "*r", ms=10)
```

El punto crítico en 13/2 es un máximo global

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x):
    return (5+x+4/x)
df=sp.diff(f(x), x)
pc=sp.solve(df, x)

ddf=sp.diff(df, x)

crit_2da_derivada=ddf.subs(x, pc[0])
if crit_2da_derivada <0:
    print("El número crítico se encuentra en "+str(pc[0]) + " y es un máximo global")
elif crit_2da_derivada >0:
    print("El punto crítico se encuentra en "+str(pc[0]) + " y es un mínimo global")
else:
    print("El punto crítico se encuentra en "+str(pc[0]) + " y no se puede definir por criterio de la segunda derivada")

print("El número crítico se encuentra en " + str(pc))
print("La segunda derivada es " + str(ddf))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(0.1,10,50)

plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    plt.plot(c, f(c), "*r", ms=10)
```

El punto crítico está en -2 y 2 pero como sólo pide valores mayores a cero el único punto crítico es 2 y se trata de un mínimo global

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

Como tenemos un perimetro de 10cm podemos decir que los dos lados iguales miden "s". 
Entonces:
2s + b = 10 cm (perímetro del triángulo)

Queremos maximizar el volumen del cono, que se da por la fórmula V = (1/3)πr^2h, donde "r" es el radio de la base y "h" es la altura del cono.

Para encontrar "r" y "h" en términos de "s" y "b", primero debemos saber que el radio "r" es igual a r = b/2, y la altura "h" es igual a "s"

Ahora debemos de poner todo en términos de "s" y "b"
 $$ r = b/2 $$
$$ h = s $$

El volumen del cono es:

$$ V = (1/3)πr^2h $$

Sustituyendo las expresiones para "r" y "h":

$$ V = (1/3)π((b/2)^2)s $$

$$ V = (1/3)π(b^2/4)s $$

1. Despejar b en términos de s
$$ b = 10 - 2s $$
 
2. Sustituyendo esto en la expresión para V:

$$ V = (1/3)π(((10 - 2s)^2)/4)s $$
3. Simplificar e igualar a cero
$$ V = (1/3)π(1/4)(100 - 40s + 4s^2)s $$

$$ V = (π/12)(100s - 40s^2 + 4s^3) $$
$$ dV/ds = (π/12)(100 - 80s + 12s^2) = 0 $$

$$ 100 - 80s + 12s^2 = 0 $$
Entonces nuestra función es:
$$ f(x)=100-80s+12s^2 $$
5. Simplificar la función
Podemos dividir entre 4 para simplificar más la expresión:
 $$ 25 - 20s + 3s^2 = 0 $$

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x):
    return (25-20*x+3*x**2)
df=sp.diff(f(x), x)
pc=sp.solve(df, x)
print(str(df))
print(str(pc))
```

Como se puede ver 10/3 es una solución pero esa sólo es "b" falta "s". Entonces utilizamos la fórmula general para encontrar los dos valores.
Y los resultados que se encuentran de la función de $$ 25 - 20s + 3s^2 = 0 $$ dan dos resultados: 5/3 y 5.
Para $$ s = 5/3 $$ $$ b = 10 - 2(5/3) = 10/3. $$



Entonces el para encontrar el volumen máximo, sustituye estos valores de "s" y "b" en la fórmula del volumen del cono:

$$ V = (1/3)π((10/3)^2)(5/3) $$

$$ V ≈ 58.85 cm³ $$ 

Así que, el volumen máximo del cono generado es aproximadamente 58.85 cm³.

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

1. La portería de fútbol tiene la forma de un rectángulo con dos postes verticales y un travesaño horizontal.
Llamemos "x" a la longitud de los postes verticales y "y" a la longitud del travesaño horizontal. 
Por lo tanto tenemos: 
$$ 2x + y = 6 $$

Podemos resolver este problema utilizando el método de sustitución. Primero, despejamos "y" de la ecuación 2x + y = 6:

$$ y = 6 - 2x $$

Luego, sustituimos esta expresión en la ecuación del área A = xy:

$$ A = x(6 - 2x) $$

$$ A = 6x - 2x^2 $$
2. Igualar a cero la función
Para maximizar el área A, tomamos la derivada de A con respecto a "x" y la igualamos a cero para encontrar los valores críticos:

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x):
    return (6*x-2*x**2)
df=sp.diff(f(x), x)
pc=sp.solve(df, x)
print(str(df))
print("Los puntos crítcos son " + str(pc))
```

Ahora que tenemos el valor de "x", podemos encontrar el valor de "y" usando la ecuación original 2x + y = 6:

$$ 2(3/2) + y = 6 $$

$$ 3 + y = 6 $$

$$ y = 6 - 3 $$

$$ y = 3 $$ 

Para maximizar el área de la portería, los postes deben medir 3/2 metros (1.5 metros) y el travesaño horizontal debe medir 3 metros. Esto dará como resultado un área máxima para la portería.
