1. Ajuste estos datos a polinomios de grado 1 hasta 3 (los datos en $x$ corresponden a los años de 2014 a 2040, sin embargo son valores muy grandes para elevarlos a una potencia, use una representación para estos años, por ejemplo $2014 \rightarrow 0$ y $2040\rightarrow 26$).

2. Muestre las tres curvas ajustadas junto a los datos de los costos en un solo gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de años (2014->0,2040->26)
años = np.arange(0,27)

# Datos de costos en dólares
costos = np.array([10.051101,  9.259127,  9.444813,  9.557723,  9.977324, 10.298326,
       10.447217, 10.550506, 10.590784, 10.657418, 10.743085, 10.831965,
       10.988435, 11.153055, 11.313326, 11.557165, 11.860136, 12.122511,
       12.356448, 12.611391, 12.814569, 13.111372, 13.357134, 13.559012,
       13.85861 , 14.135924, 14.436149])

# Ajuste polinómico de grado 1
grado_1=np.polyfit(años,costos,1)

# Ajuste polinómico de grado 2
grado_2=np.polyfit(años,costos,2)

# Ajuste polinómico de grado 3
grado_3=np.polyfit(años,costos,3)

# Funciones polinómicas de grado 1, 2 y 3
def polinomio_grado_1(x):
    return grado_1[0]*x+grado_1[1]

def polinomio_grado_2(x):
    return grado_2[0]*x**2+grado_2[1]*x+grado_2[2]

def polinomio_grado_3(x):
    return grado_3[0]*x**3+grado_3[1]*x**2+grado_3[2]*x+grado_3[3]

# Puntos para representar las curvas ajustadas
años_ajuste=np.linspace(0,26,100)
costos_grado_1=polinomio_grado_1(años_ajuste)
costos_grado_2=polinomio_grado_2(años_ajuste)
costos_grado_3=polinomio_grado_3(años_ajuste)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(años,costos,label='Datos de Costos',color='blue',marker='o')
plt.plot(años_ajuste,costos_grado_1,label='Ajuste Grado 1',color='red')
plt.plot(años_ajuste,costos_grado_2,label='Ajuste Grado 2',color='green')
plt.plot(años_ajuste,costos_grado_3,label='Ajuste Grado 3',color='purple')

# Configuración del gráfico
plt.title('Ajuste de Datos a Polinomios de Grado 1 hasta 3')
plt.xlabel('Años (Desde 2014 hasta 2040)')
plt.ylabel('Costo en Dólares')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de años y población
años=np.array([1950, 1951, 1952, 1953, 1954, 1955, 1956, 1957, 1958, 1959, 1960,
                 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968, 1969, 1970, 1971,
                 1972, 1973, 1974, 1975, 1976, 1977, 1978, 1979, 1980, 1981, 1982,
                 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993,
                 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004,
                 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])

poblacion=np.array([28296, 29110, 29980, 30904, 31880, 32906, 33978, 35095, 36253, 37448,
                      38677, 39939, 41234, 42564, 43931, 45339, 46784, 48264, 49788, 51361,
                      52988, 54669, 56396, 58156, 59931, 61708, 63486, 65261, 67013, 68715,
                      70353, 71916, 73416, 74880, 76351, 77859, 79410, 80999, 82635, 84327,
                      86077, 87890, 89758, 91654, 93542, 95393, 97202, 98969, 100679, 102317,
                      103874, 105340, 106724, 108056, 109382, 110732, 112117, 113530, 114968,
                      116423, 114255, 115683, 117054, 118395, 119713, 121005])

# Ajuste polinómico de grado 1 hasta 3
grado_1=np.polyfit(años,poblacion, 1)
grado_2=np.polyfit(años,poblacion, 2)
grado_3=np.polyfit(años,poblacion, 3)

# Funciones polinómicas de grado 1, 2 y 3
def polinomio_grado_1(x):
    return grado_1[0]*x+grado_1[1]

def polinomio_grado_2(x):
    return grado_2[0]*x**2+grado_2[1]*x+grado_2[2]

def polinomio_grado_3(x):
    return grado_3[0]*x**3+grado_3[1]*x**2+grado_3[2]*x+grado_3[3]

# Puntos para representar las curvas ajustadas
años_ajuste=np.linspace(1950, 2015, 100)
poblacion_grado_1=polinomio_grado_1(años_ajuste)
poblacion_grado_2=polinomio_grado_2(años_ajuste)
poblacion_grado_3=polinomio_grado_3(años_ajuste)

# Crear la gráfica
plt.figure(figsize=(10,6))
plt.scatter(años,poblacion,label='Datos de Población',color='blue',marker='o')
plt.plot(años_ajuste,poblacion_grado_1,label='Ajuste Grado 1',color='red')
plt.plot(años_ajuste,poblacion_grado_2,label='Ajuste Grado 2',color='green')
plt.plot(años_ajuste,poblacion_grado_3,label='Ajuste Grado 3',color='purple')

# Título de la gráfica
plt.title('Ajuste de Datos de Población a Polinomios de Grado 1 hasta 3')
plt.xlabel('Años (1950-2015)')
plt.ylabel('Población')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

Luego para graficar el error cuadrático acumulado contra el número de términos es necesario:

```python
# Calcular el error cuadrático 
errores_cuadraticos=[]

for grado in range(1, 4):
    coeficientes=np.polyfit(años, poblacion, grado)
    y_pred=np.polyval(coeficientes, años)
    error=np.sum((poblacion-y_pred)**2)
    errores_cuadraticos.append(error)

# Graficar el error cuadrático 
plt.figure(figsize=(8,5))
plt.plot(range(1, 4), errores_cuadraticos, marker='o', linestyle='-')
plt.title('Error Cuadrático Acumulado vs. Grado del Polinomio')
plt.xlabel('Grado del Polinomio')
plt.ylabel('Error Cuadrático Acumulado')
plt.grid(True)
plt.show()
```
